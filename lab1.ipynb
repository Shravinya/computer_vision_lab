```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread(r"C:\Users\SHRAVINYA\Downloads\lab1.png")

# Set blue channel (index 0) to 0
img[1,:,: ] = 0

# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image without blue
plt.imshow(img_rgb)
plt.title("Blue Channel Removed (B = 0)")
plt.axis("off")
plt.show()
```
