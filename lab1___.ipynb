 ## Color Isolation

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image_path = r"C:\Users\SHRAVINYA\Downloads\strawberry.jpg"
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(img_hsv)

# Split RGB channels
r, g, b = cv2.split(img_rgb)

# Create RGB channel variations
only_r = np.zeros_like(img_rgb)
only_r[:, :, 0] = r
only_g = np.zeros_like(img_rgb)
only_g[:, :, 1] = g
only_b = np.zeros_like(img_rgb)
only_b[:, :, 2] = b

# Plotting
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes[0, 0].imshow(img_rgb)
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')
axes[0, 1].imshow(h, cmap='hsv')
axes[0, 1].set_title('Hue Channel')
axes[0, 1].axis('off')
axes[0, 2].imshow(s, cmap='gray')
axes[0, 2].set_title('Saturation Channel')
axes[0, 2].axis('off')
axes[1, 0].imshow(v, cmap='gray')
axes[1, 0].set_title('Value Channel')
axes[1, 0].axis('off')
axes[1, 1].imshow(only_r)
axes[1, 1].set_title('Red Channel Only')
axes[1, 1].axis('off')
axes[1, 2].imshow(only_g)
axes[1, 2].set_title('Green Channel Only')
axes[1, 2].axis('off')
axes[2, 0].imshow(only_b)
axes[2, 0].set_title('Blue Channel Only')
axes[2, 0].axis('off')
axes[2, 1].axis('off')
axes[2, 2].axis('off')
plt.tight_layout()
plt.show()
```
